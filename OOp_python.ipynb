```python
class Car:
    def __init__(self ,brand,color):
        self.brand=brand
        self.color=color
    def show(self):
        print(f"Car Brand: {self.brand}, Color: {self.color}")


car=Car('BMW','Black')
car.show()
```

```python
class Student:
    def __init__(self,age,name,gender):
        self.age=age
        self.name=name
        self.gender=gender
    def show_info(self):
        print(f"Age: {self.age}, Name: {self.name}, Gender: {self.gender}")

student=Student(20,"John", "Male")
student2=Student(21,"John", "Male")
student3=Student(22,"Hamza", "Male")
student.show_info()
student2.show_info()
student3.show_info()
```

```python
class Mobile:
    def __init__(self,brand,price):
        self.brand=brand
        self.price=price
    def show(self):
        print(f"Mobile Brand: {self.brand}, Price: {self.price}")

mobile=Mobile('Samsung',1000)
mobile.show()
```

```python
class Laptop:
    def __init__(self,brand,ram):
        self.brand=brand
        self.ram=ram
    def show(self):
        print(f"Laptop Brand: {self.brand}, Ram: {self.ram}")

laptop1=Laptop('Hp','8GB')
laptop2=Laptop('Dell','8GB')
laptop3=Laptop('Hp','16GB')
laptop4=Laptop('Dell','16GB')
laptop1.show()
laptop2.show()
laptop3.show()
laptop4.show()
```
