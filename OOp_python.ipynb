```python
class ATM:
    def __init__(self,initial_balance=0,interest_rate=0.01,transactions=[],account_number=""):
        self.balance=initial_balance
        self.transactions=[]
        self.account_number=account_number
    def deposit(self,amount):
        self.balance+=amount
        self.transactions.append(f"Deposited {amount}")
    def withdraw(self,amount):
        if self.balance>=amount:
            self.balance-=amount
            self.transactions.append(f"Withdrew {amount}")
        else:
            print("Insufficient balance")
    def get_balance(self):
        return self.balance
    def get_transactions(self):
        return self.transactions
#here we create obj

obj=ATM(1000)
obj.deposit(500)
obj.withdraw(200)
print(obj.get_balance())
print(obj.get_transactions())
```

```python
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result

    def subtract(self, num):
        self.result -= num
        return self.result

    def multiply(self, num):
        self.result *= num
        return self.result

    def divide(self, num):
        if num != 0:
            self.result /= num
            return self.result
        else:
            return "Cannot divide by zero"

# Example usage:
calc = Calculator()
print(calc.add(5))        # Output: 5
print(calc.subtract(3))   # Output: 2
print(calc.multiply(4))   # Output: 8
print(calc.divide(2))     # Output: 4
print(calc.divide(0))     # Output: Cannot divide by zero
```
