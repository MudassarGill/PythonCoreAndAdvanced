```python
class InsufficientBalance(Exception):
    pass

balance = 500

try:
    withdraw = int(input("Enter withdraw amount: "))

    if withdraw > balance:
        raise InsufficientBalance("Not enough balance")

    balance -= withdraw
    print("Remaining balance:", balance)

except InsufficientBalance as e:
    print("Error:", e)
```
