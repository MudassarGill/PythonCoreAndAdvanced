```python
import threading
```

```python
def print_numbers():
    for i in range(1, 6):
        print(f"Number: {i}")
```

```python
thread1 = threading.Thread(target=print_numbers)
thread1.start()  # Thread start
thread1.join()   # Wait until thread finishes
```

```python
import threading
import time

def task(name, delay):
    for i in range(3):
        time.sleep(delay)
        print(f"{name} task iteration {i+1}")

# Create threads
thread1 = threading.Thread(target=task, args=("Thread-1", 1))
thread2 = threading.Thread(target=task, args=("Thread-2", 2))

# Start threads
thread1.start()
thread2.start()

# Wait for threads to finish
thread1.join()
thread2.join()

print("All tasks completed")
```

```python
from multiprocessing import Process
```

```python
def square(n):
    print(f"Square of {n} = {n*n}")
```

```python
p1 = Process(target=square, args=(5,))
p1.start()  # Process start
p1.join()   # Wait until process finishes
```

```python
from multiprocessing import Process
import time

def task(name, delay):
    for i in range(3):
        time.sleep(delay)
        print(f"{name} iteration {i+1}")

# Create processes
p1 = Process(target=task, args=("Process-1", 1))
p2 = Process(target=task, args=("Process-2", 2))

# Start processes
p1.start()
p2.start()

# Wait for completion
p1.join()
p2.join()

print("All processes completed")
```
