```python
import threading
```

```python
def print_numbers():
    for i in range(1, 6):
        print(f"Number: {i}")
```

```python
thread1 = threading.Thread(target=print_numbers)
thread1.start()  # Thread start
thread1.join()   # Wait until thread finishes
```

```python
import threading
import time

def task(name, delay):
    for i in range(3):
        time.sleep(delay)
        print(f"{name} task iteration {i+1}")

# Create threads
thread1 = threading.Thread(target=task, args=("Thread-1", 1))
thread2 = threading.Thread(target=task, args=("Thread-2", 2))

# Start threads
thread1.start()
thread2.start()

# Wait for threads to finish
thread1.join()
thread2.join()

print("All tasks completed")
```
