```python
def largest_number(a,b,c):
    if a>=b and b>=c:
        return a
    elif b>=a and b>=c:
        return b
    else:
        return c

largest_number(10,50,3)
```

```python
def factorial(n):
    result=1
    for i in range(1,n+1):
        result=result*i
    return result

factorial(5)
```
